```python
nlon = 7                # Number of longitute values
nlat = 5                # Number of latitude values
dlon, dlat = 5.0, 5.0   # Increment

# Arrays
d = (np.arange(nlon * nlat) - (nlon * nlat / 2.0)) ** 2.0
d.shape = (nlat, nlon)

# Plot graph using matplotlib
fig = plt.figure()                 
ax = fig.add_subplot(111)         
im = ax.imshow(d, origin='lower')  
plt.colorbar(im)                   
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Coordinates of MCOM
mcom_lon = -101.8803718553
mcom_lat = 33.5818617015

# Calculate lon and lat arrays
lon = np.linspace(mcom_lon - dlon * (nlon - 1) / 2, mcom_lon + dlon * (nlon - 1) / 2, nlon)
lat = np.linspace(mcom_lat - dlat * (nlat - 1) / 2, mcom_lat + dlat * (nlat - 1) / 2, nlat)

# Set custom tick labels for x and y axes with two decimal places and increased spacing
x_ticks_positions = np.linspace(0, len(lon) - 1, len(lon))
ax.set_xticks(x_ticks_positions)
ax.set_xticklabels([f'{x:.2f}' for x in lon], rotation=45)  # Rotating the labels for better visibility
ax.set_yticks(np.arange(len(lat)))
ax.set_yticklabels([f'{y:.2f}' for y in lat])

# Set the title of the plot
ax.set_title('2D Data with 1D Arrays (Centered on MCOM)')

print("lon:",lon)
print("lat:", lat)

plt.show()
```
