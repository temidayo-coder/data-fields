**3.** Write a function that converts 1D center positions to edge positions. The function should calculate the grid spacing using only the information in the variable `x`. Use the closest adjacent spacing to extend the start and end of the array. Use this function to create two arrays `lat_edge` and `lon_edge` and print their values. Is the new figure created with `imshow` correct? Explain your answer. (20 pts.)

### Solution to Question 3:  
#### Cleaning it up

```python
def centers_to_edges_1d(x):
    
    # replace with your code. See code below.

    # Compute the grid spacing by taking the average of adjacent spacings and store in the variable "spacing"
    spacing = np.mean(np.diff(x))
    
    # Extend the array at the start and end, using 0.5 as the closest adjacent spacing
    start = x[0] - 0.5 * spacing
    end = x[-1] + 0.5 * spacing
    
    # Create the edge positions array
    edges = np.arange(start, end + spacing, spacing)
    
    return edges

# Test the function with lon and lat arrays 
lon_edge = centers_to_edges_1d(lon)
lat_edge = centers_to_edges_1d(lat)
print("lon_edge:", lon_edge)
print("lat_edge:", lat_edge)


# Now let's visualize our new plot
fig = plt.figure()
ax = fig.add_subplot(111)
im = ax.imshow(d, origin='lower', extent=(lon_edge.min(), lon_edge.max(), lat_edge.min(), lat_edge.max()))
plt.colorbar(im)
ax.scatter(mcom_lon, mcom_lat, marker='x', color='red')
ax.set_xlabel('longitude (degrees)')
ax.set_ylabel('latitude (degrees)')

# For clarity, I will add dashed lines to indicate the exact position of MCOM
ax.axvline(x=mcom_lon, color='gray', linestyle='--')
ax.axhline(y=mcom_lat, color='gray', linestyle='--')

# Also, let's display the latitude and longitude values of the dashed lines using .text function
ax.text(mcom_lon + 1, lat_edge.min(), f'Lon: {mcom_lon:.2f}', color='gray')
ax.text(lon_edge.min(), mcom_lat + 1, f'Lat: {mcom_lat:.2f}', color='gray')
```
